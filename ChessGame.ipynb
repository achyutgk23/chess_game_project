# Chess Game

Chess is a board game which can be played between two players. Chess is a game of patterns and stratergies. A typical chess game played between two professionals can be divided into three phases viz. 'opening game', 'middle game' and 'end game' and for each of these phases powerful strategies has to be made in order to win a game. 

Traditionally chess is a physical board game played between two actual people, but the online chess has revolutionized the game as any two people sitting in different corners of the world can play a chess game togeher. The advent of online chess game has certainly pushed the boundaries and limitations of a physical board game of chess and it will only help increase the popularity of chess game even further. 

It is already universal that the game of chess is the game of intelligence, strategy, sophistication and complexity. The planning, thinking and strtergies used, and being developed in chess, have relevence to the actual world and they can be applied in business, politics, wars etc. 

## Business Understanding

Obviously the player who wins the touranament wins the entitled cash prize but the main important thing is he gets recognition among the community which is very important for his career point of view and to build a brand around himself for monetory gains in the future.

1. What are the major ways of obtaining each outcome of a game?

2. What are the top 10 most used increment codes and which outcome of the game is dominating each of the top 10 increment codes?

3. How many turns were played, on an average, in each of the top 10 increment codes?

4. What are the top 10 most used opening games and which outcome of the game is dominating each of the top 10 openings?

5. How well can we predict an outcome of a game?

## Data Understanding

The data has 16 columns and 20058 observations.

The variables in the dataset are as follows:
1.	id – Unique Identification Number of a game.

2.	rated – Whether a game is rated or casual.

3.	created_at – The time recorded when the game has started.

4.	last_move_at – The time recorded when the game has finished.

5.	turns – Number of turns in a game.

6.	victory_status – In what way the victory or an outcome of a game is obtained and there are four types of victory_status viz. ‘check mate’, ‘out of time’, ‘resign’, and ‘draw’.

7.	winner – Outcome of a game i.e., black piece player’s win, white piece player’s win and draw.

8.	Increment_code – Increment code is time expressed in the form ‘5+8’ which means the total time for each player is 5 minutes and for every move by a player, extra 8 seconds are added to the player's total time.

9.	white_id – ID of white player.

10.	white_rating – Rating of white player.

11.	black_id – ID of black player.

12.	black_rating – Rating of black player.

13.	moves – Every move in a game by both players (Scorecard of a game).

14.	opening_eco – Encyclopedia of Chess Openings which describes and classifies the openings of a chess game.

15.	opening_name – The name of the opening in a chess game.

16.	opening_ply – The number of turns played by both players in an opening game.

### 1. What are the major ways of obtaining each outcome of a game?

As we already know, each outcome of a game can be obtained in many different ways. A win for either player can be obtained in three ways viz. 'resign', 'checkmate' and 'out of time'. A draw can happen in two ways viz. 'out of time' and 'draw' (through stalemate, mutual agrrement, threefold repitition etc). Here we are trying to find out the major way of winning for each of the outcomes of a game.

```python
# Creaing the dataframe
vic_stat_df = df.groupby(['winner', 'victory_status']).count().id.reset_index(name='count')

# Selecing index and columns of the dataframe
vic_stat_df = vic_stat_df.pivot(index='victory_status', columns='winner', values='count')

# Filling NA values with 0
vic_stat_df.fillna(0, inplace=True)

vic_stat_df
```

```python
# Printing the total occurrence of each outcome of a game
print(df['winner'].value_counts())

# Visualizing with a heatmap
sns.heatmap(vic_stat_df, annot=True, fmt='.4g', robust=False, cmap="Paired");
title('Major Ways of obtaing each outcome of a game')
```

### 2. What are the top 10 most used increment codes and which outcome of the game is dominating each increment codes?

#### Top 10 most used increment codes

```python
# Creating the dataframe of top 10 most used increment codes in the dataset,
# resetting the index of the dataframe and renaming the columns
top10_incr_codes = df["increment_code"].value_counts().sort_values(ascending = False)[:10]
top10_incr_codes_df = pd.DataFrame(top10_incr_codes).reset_index()
top10_incr_codes_df.rename(
    columns={'index':'increment_code', 'increment_code':'count'},
    inplace=True
)

top10_incr_codes_df
```

```python
# Plotting the top 10 most used increment codes
top10_codes_plot = sns.catplot(data=top10_incr_codes_df, kind='bar', 
                               x='increment_code', y='count', palette='dark',
                               aspect=1.2);
text_in_plot(top10_codes_plot.ax)
title('Top 10 most used increment codes')
```

#### Which outcome of a game dominates each increment code

```python
# Identifying which outcome of the game (i.e. white, black and draw) has more
# number of occurance in each of the top 10 increment codes

# Creating a list of top 10 increment codes
incr_names = list(top10_incr_codes.index[:10])

# Empty dictionary for storage
incr_dict = {}

# Writing a for loop which gives number of different outcomes for each increment codes
for number in range(10):
    incr_dict[incr_names[number]] = df[
        df["increment_code"]==incr_names[number]]["winner"].value_counts()
    
# concatenating incr_dict dictionary along 0 axis (along rows)
incr_sr = pd.concat(incr_dict, axis=0)

# Converting series into dataframe, resetting index and renaming the columns
incr_df = pd.DataFrame(incr_sr).reset_index()
incr_df.rename(columns={'level_0':'increment_code',
                       'level_1':'colour',
                       'winner':'count'}, inplace=True)

incr_df
```

```python
# Plotting total number of wins by all 3 outcomes of the game (black, white and draw)
# in each of the top 10 increment codes
incr_plot = sns.catplot(data=incr_df, kind='bar', 
            x='increment_code', y='count', 
            hue='colour', aspect=1.8, palette='muted');
text_in_plot(incr_plot.ax)
title('Outcome of a game that dominates each increment codes')
```

### 3. How many turns were played in all top 10 most used increment codes, on average?

```python
# Claculating average turns for all the top 10 increment codes
avg_turns_incr_df = pd.DataFrame(
    df.groupby(['increment_code']).mean().turns.round()).reset_index()
avg_turns_incr_df
```

```python
# Out of all the increment codes, extracting only the top 10
# most used increment codes along with their average turns
incr_turns_dict={}
for i in range(10):
    incr_turns_dict[i] = avg_turns_incr_df[
        avg_turns_incr_df['increment_code']==incr_names[i]]
turns_top_incr_df = pd.concat(incr_turns_dict, axis=0)
turns_top_incr_df.reset_index(inplace=True)
turns_top_incr_df.drop(['level_0', 'level_1'], axis=1, inplace=True)

turns_top_incr_df
```

```python
# Visualizing with point plot
sns.catplot(data=turns_top_incr_df,
            x='increment_code', y='turns',
            kind='point', palette='dark',
            aspect=1.3);
title('Average number of turns played in top 10 most used increment codes')
```

### 5. How well can we predict an outcome of a game?

#### Checking for Missing Values in the Dataset

#### Model Selection

```python
# From the main data set extracting only the numerical variables for model selection
df_num = df.select_dtypes(include=['int64', 'float64', 'bool']).copy()

# Adding the target variable to numerical dataset
df_num['winner'] = df['winner'].copy()

# Creating X and y dataframes for training and testing
X_num = df_num.drop('winner', axis=1)
y_num = df_num['winner']
```

```python
# Splitting the data into training set and test set
from sklearn.model_selection import train_test_split

X_num_train, X_num_test, y_num_train, y_num_test = train_test_split(X_num, 
                                                                    y_num, 
                                                                    test_size=0.3, 
                                                                    random_state=12)
```

```python
# Feature scaling the training set so that all the numeric variables
# should have same scale
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Transforming the training set with StandardScaler
X_num_train_scaled = scaler.fit_transform(X_num_train.astype(np.float64))
```

```python
# Creating a function which takes a classifier, variables in X matrix and
# corresponing labels from y matrix as inputs, generate cross validated estimates
# for each input data points and shows the scores of different
# evaluation metrics like accuracy, precision, recall

def model_selection(classifier, X_train, y_train):
    '''
    INPUT:
    classifier - From the list of classifiers considering one classifier at a time
                 to evaluate its performance on validation sets(created by training set)
    X_train - Pandas Dataframe with predictors/variables
    y_train - Pandas Series with corresponding labels
    
    OUTPUT:
    1. A classifier from the list of classifiers predicts on 
       different validation sets and the predictions are stored in preds variable
    2. Prints scores of different evaluation metrics like
        a. Accuracy
        b. Precision
        c. Recall
        and prints the label from y_train which the classifier did not 
        predict
    '''
    
    from sklearn.metrics import precision_score, recall_score, accuracy_score
    from sklearn.model_selection import cross_val_predict
    
    classifier.fit(X_train, y_train)
    preds = cross_val_predict(classifier, X_train, y_train, cv=3,
                          n_jobs=-1)
    print('accuracy:', accuracy_score(y_train, preds))
    print('precision:', precision_score(y_train, preds, average='weighted'))
    print('recall:', recall_score(y_train, preds, average='weighted'))
    print(set(y_train) - set(preds))
```

#### Evaluation Metrics used for anayzing the performance of classifiers

1. **Accuracy**: Accuracy is the ratio of correct predictions to the total predictions i.e., it tells us the proportion of correct predictions among all the predictions. 


2. **Precision**: It is the ratio of predictions which are correctly predicted as positives by the classifier to all the predictions which are predicted as positives by the classifier i.e., it gives us the proportion of correct positive predictions amongst all the positive predictions.


3. **Recall**: It is also called as 'true positive rate' and 'sensitivity'. It is the ratio of predictions which are correctly predicted as positives by classifier to the actual positives i.e., it gives fraction of correct positive predictions amongst all the actual positives.

```python
# Importing required libraries of different classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier

# Instantiating the different classifiers
log_reg = LogisticRegression()
sdg_clf = SGDClassifier()
softmax_reg = LogisticRegression(
    multi_class="multinomial",
    solver="lbfgs", C=10)
tree_clf = DecisionTreeClassifier()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)
knn_clf = KNeighborsClassifier()
voting_clf = VotingClassifier(
     estimators=[('lr', log_reg), 
                 ('rf', rnd_clf), 
                 ('svm_clf', svm_clf), 
                 ('tree_clf_train', tree_clf),
                ('knn', knn_clf)],
     voting='hard')

# Creating a list of all the instantiated classifiers
classifiers = [log_reg,
              sdg_clf,
              softmax_reg,
              tree_clf,
              rnd_clf,
              svm_clf,
              knn_clf,
              voting_clf]

# Writing a for loop which takes each of the classifiers
# and using model_selection function, prints the accuracy,
# precision and recall for each classifier on validation sets created by training set
for clf in classifiers:
    print(clf)
    model_selection(clf, X_num_train_scaled, y_num_train)
    print()
```

Based on the performance of all the classifiers, it is evident that Random Forest Classifier and Voting Classifier have performed better than other classifiers. Out of the above mentioned two classifiers, Random Forest Classifier is selected, out of convinience for the hyper parameter tuning, for making further predictions

#### Using Categorical Variables to Predict the Outcome of the Game with Random Forest Classifier

```python
# Now along with numeric variables, we inlcude categorical variables 
# to predict the winner (outcome) of a game

# Creating a dataframe of categorical variables from the main dataframe
df_cat = df.select_dtypes(exclude=['int64', 'float64', 'bool']).copy()

# Dropping the columns 'id, 'white_id, 'black_id', 'moves' and 'winner'.
# because of the problem of too many variables which are greater than the
# number of observations
# Dropping the 'winner' column as it is the target variable
df_cat.drop(['id', 'white_id', 'black_id', 'moves', 'winner'], 
            axis=1, inplace=True)

# Creating the dummy variables of categorical variables and dropping the
# first column so that we get k-1 number of variables
df_cat_dm = pd.get_dummies(df_cat, drop_first=True)

# After creating dummy variables concatenating both dataframes with numeric
# and categorical variables
df_final = pd.concat([df_num, df_cat_dm], axis=1)
```

```python
# Creating the X and y variables for training and testing
X = df_final.drop(['winner'], axis=1)
y = df_final['winner']

# Splitting the X and y matrices into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3,
                                                   random_state=23)

# Feature scaling the training set of predictors
X_train_scale = scaler.fit_transform(X_train.astype('float64'))

# Finding the performance of random forest classifier on validation sets
# using previously defined 'model_selection' function
model_selection(rnd_clf, X_train_scale, y_train)
```

#### Fine Tuning the Hyperparameters of the Random Forest Classifier

```python
# Importing GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Creating a parameter grid for hyperparameter tuning
param = [
    {'n_estimators': [3, 10, 100],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 4, 6],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [4, 8, 12]}
]

# Using GridSearchCV, instantiating the random forest classifier
# along with parameter grid and fitting the classifier to training set
clf = GridSearchCV(rnd_clf, param)
clf.fit(X_train_scale, y_train)
```

```python
# Storing the best model
final_model = clf.best_estimator_
final_model

# Making predictions on feature scaled X_train
preds_train = final_model.predict(X_train_scale)
```

```python
#Creating 'show_scores' function which displays the various evaluation
# matrices like accuracy, precision, recall and confusion matrix
def show_scores(y_train, preds):
    '''
    INPUT:
    y_train - A pandas series with target variable meant for
              training a classifier
    preds - Predictions made by a classifier on training set (X_train)
    
    OUTPUT:
    Outputs scores of different evaluation metrics for a classifier like
        1. Accuracy
        2. Precision
        3. Recall
        4. Confusion matrix
    '''
    
    from sklearn.metrics import accuracy_score, confusion_matrix
    from sklearn.metrics import precision_score, recall_score
    
    print('accuracy:', accuracy_score(y_train, preds))
    print('precision:', precision_score(y_train, preds, average='weighted'))
    print('recall:', recall_score(y_train, preds, average='weighted'))
    print('confusion_matrix:')
    print(confusion_matrix(y_train, preds))
    
# Using the 'show_scores' function to evaluate the performance
# of hyper tuned random forest classifier on training set
show_scores(y_train, preds_train)
```

#### Predictions on the Test Set

```python
# Feature scaling the X_test for prediction by fine tuned random forest
# classifier
X_test_scaled = scaler.fit_transform(X_test.astype(np.float64))

# Making predictions on feature scaled X_test
preds_test = final_model.predict(X_test_scaled)

# Using the 'show_scores' function to evaluate the performance
# of hyper tuned random forest classifier on test set
show_scores(y_test, preds_test)
```
